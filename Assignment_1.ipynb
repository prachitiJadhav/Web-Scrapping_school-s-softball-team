```python
from bs4 import BeautifulSoup
import requests

url_1 = "https://eastbaypioneers.com/sports/softball/roster"
text_1 = requests.get(url_1).text
soup = BeautifulSoup(text_1, "html5lib") 
print(soup.prettify())





```

```python
all_urls_1 = [a['href'] for a in soup('a')  if a.has_attr('href')]
all_urls_1
print(len(all_urls_1))
print(all_urls_1)
```

```python
import re
regex = r"^/sports/softball/roster/.*$"
# this requires the URL must start with ^/sports/softball/roster/.*$

# we can test some cases.
assert re.match(regex, "/sports/softball/roster/mariah-ayala/4545")
assert not re.match(regex, "https://eastbaypioneers.com/sports/softball/roster?print=true")

# apply to our original list
useful_urls = [l for l in all_urls_1 if re.match(regex,l)]
print(len(useful_urls))
print(useful_urls)
```

```python
#remove duplicates
useful_urls = list(set(useful_urls))
print(useful_urls)
print(len(useful_urls))
```
